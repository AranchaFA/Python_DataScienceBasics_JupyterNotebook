## INSTALANDO Jupyter Notebook CON Anaconda

Instalaremos el paquete Jupyter Notebook en nuestro entorno Python utilizando Anaconda.

Anaconda funciona como gestor de paquetes para entornos Python (Python environments). Permite crear distintos entornos Python e instalar en cada uno de ellos los paquetes que deseemos, para tener entornos de trabajo separados con caractersticas diferentes.
Podemos instalar Anaconda si en nuestro equipo ya tenemos instalado Python, tambin existen versiones de Python para descargar que ya incorporan Anaconda.

Anaconda tiene su propia consola (es importante acordarse de no ejecutar los comandos en la consola del sistema, como la cmd de Windows!). 

Los comandos que necesitamos para instalar Jupyter Notebook (y algunos otros bsicos que nos pueden ser de ayuda):

- **Crear un entorno en anaconda**:

    - \$ conda create -n "nombre_del_entorno" python="python_version"
    
- **Activar un entorno en anaconda**:

    - \$ conda activate "nombre_del_entorno"

- **Desactivar un entorno en anaconda**:

    - ("nombre_del_entorno") \$ conda deactivate
    
- **Instalar un paquete o librera en anaconda** (debemos tener el entorno activado)

    - ("nombre_del_entorno") \$ conda install "nombre_del_paquete"
    
- **Activar un entorno de anaconda para usarlo dentro de un jupyter notebook** (debemos tener el entorno activado y ipykernel instalado)

    - ("nombre_del_entorno") \$ python -m ipykernel install --user --name "nombre_del_entorno" --display-name "Nombre en Jupyter"
    
- **Lanzar un jupyter notebook** (debemos tener el entorno activado y el paquete jupyter notebook instalado dentro)

    - ("nombre_del_entorno") \$ jupyter notebook
    
- **Exportar las libreras instaladas en el entorno a un archivo txt**

    - ("nombre_del_entorno") \$ conda export > "nombre_del_archivo".txt
    
- **Instalar las libreras desde un archivo .txt**

    - ("nombre_del_entorno") \$ conda install -r "nombre_del_archivo".txt
    
- **Ver las libreras instaladas dentro de un entorno**

    - ("nombre_del_entorno") \$ conda list
    
- **Ver todos los entornos que tengo creados en mi ordenador**

    - \$ conda env list

## Shortcuts de Jupyter Notebook

Una celda con barra **azul** a la izquierda est en **"Command Mode"**, en este modo tenemos los shortcuts:

- **M**: cambia la celda a tipo "Markdown" (para escribir texto)
- **Y**: cambia la celda a tipo "Code" (para escribir cdigo)
- **Up_arrow**/**Down_Arrow**: selecciona la celda de arriba/abajo
- **Ctrl + Enter**: corre el cdigo o el texto que haya dentro de la celda
- **Shift + Enter**: corre el cdigo o el texto que haya dentro de la celda y pasa a la siguiente celda
- **A**: (Above) crea una celda vacia justo encima
- **B**: (Below) crea una celda vaca justo debajo
- **D + D**: (Pulsar dos veces la tecla D) elimina la celda en la que nos encontramos

Si clicamos dentro de la celda se vuelve **verde** y pasa a **"Edit Mode"**, en este modo tenemos los shortcuts:

- **Tab**: Mientras escribimos cdigo nos sugiere o autocompleta el cdigo / Indentacin
- **Shift + Tab**: Mientras escribimos cdigo nos lanza una ventana de ayuda (funciona slo en ciertas ocasiones) / Indentacin
- **Ctrl + Enter**: corre el cdigo o el texto que haya dentro de la celda
- **Shift + Enter**: corre el cdigo o el texto que haya dentro de la celda y pasa a la siguiente celda
- **Ctrl + Z**: Undo
- **Ctrl + Y**: Redo
